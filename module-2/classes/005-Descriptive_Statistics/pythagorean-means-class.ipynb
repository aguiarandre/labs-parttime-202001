# Applications

# F1 Score

$\rightarrow$ Importance: DO NOT let zeroes take place.

$$F_1 = \frac{2}{\frac{1}{x} + \frac{1}{y}}$$

When x (or y) tends to zero, $\frac{1}{x}$ (1 divided by a really small number) tends to $\infty$ and, hence, $F_1 = \frac{2}{\infty}$ (2 divided by a huge number) tends to 0.

Imagine you have 2 important metrics and both needs to be taken into account (in machine learning you will see the example of `precision` and `recall`). 

Imagine now that you have on one side x=0.95 (a very good score) and, on the other side y= 0.1 (a complete horrible score). 
Say you use the arithmetic mean instead. The average would lead to $\overline{x} = \frac{0.95 + 0.1}{2} = 0.525$, which is a reasonable value. 
However, using the <b>harmonic mean</b>, the result you'd obtain would be:

> $F_1 = \frac{2}{\frac{1}{0.95} + \frac{1}{0.1}} = \frac{2}{1.05263158 + 10} = \frac{2}{1.05263158 + 10} = \frac{2}{10.05263158} \approx 0.2 $
