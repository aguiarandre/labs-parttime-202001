# Functional Paradigm Intro

What other paradigms we are used to?

> <b> Procedural Programming </b>
- Instructions are procedures.
- Side effects is its core.

> <b> Objected Oriented Programming </b>
- Instructions are grouped as part of a state of an object.

> <b> Functional Programming </b>
- No state exists. Just a serie of functions being evaluated. 
- No side effects.
- The solution obtained is entirely based on the input. Like in math where <code>f(x) = y</code>
- This idea leads to the fact that you can also <b>pass functions as arguments</b>. And this helps a lot.

# Map vs For-loops

```python
big_list = list(range(1000))
```

```python
def do_something(x):
    return x * 10 + 2
```

```python
%%timeit

results = []
for item in big_list:
    results.append(do_something(item))
```

```python
%%timeit

results = map(do_something, big_list)
```

```python
%%timeit

results = list(map(do_something, big_list))
```

```python
%%timeit

results = [do_something(item) for item in big_list]
```

```python
results = map(do_something, big_list)
```

```python
big_list[:10]
```

```python
list(results)[:10]
```

```python
%%timeit

results = [item * 10 + 2 for item in big_list]
```

```python
def do_something_complex(x):
    y = x * 10 + 2
    z = np.log(y)
    w = abs(z) ** 2
    
    return w
```

```python
from tqdm.auto import tqdm
```

```python
results = []
for item in tqdm(big_list):
    results.append(do_something_complex(item))
```

```python
%%timeit

results = map(do_something_complex, big_list)
```

```python
%%timeit

results = list(map(do_something_complex, big_list))
```

```python
%%timeit

results = [do_something_complex(item) for item in big_list]
```

```python
results = [do_something_complex(item) for item in big_list]
```
