```python
def correct_species(x, shark_list):
    
    if re.findall('tiger shark', x, flags=re.IGNORECASE):
        return 'Tiger Shark'
    if re.findall('white shark', x,flags=re.IGNORECASE):
        return 'White Shark'
    else:
        return x
    
```

```python
def correct_species(x, shark_list):
    
    for shark in shark_list:    
        if re.findall(shark, x, flags=re.IGNORECASE):
            return shark.capitalize() + ' Shark'
    
    return x
```

```python
df['species_treated'] = df.species.fillna('Invalid').apply(correct_species, shark_list=shark_list)
```

```python
date_list = []


bar = tqdm(df.href)

for url in bar:
    bar.set_description(url.split('/')[-1])
    headers = {'User-Agent':np.random.choice(user_agent_list)}
    try:
        response = requests.get(url, headers=headers, allow_redirects=True)
    except ConnectionError:
        print(f'ConnectionError for {url}')
        continue
        
    tmp_file = url.split('/')[-1]
    
    with open(tmp_file, 'wb') as file:
        file.write(response.content)
    
    time.sleep(5)
    
    try:
        content = textract.process(tmp_file)
    except:
        print(f'Could not read from {url.split("/")[-1]}')
    finally:
        os.remove(tmp_file)
    
    text = content.decode()
    date_list.append(re.findall('DATE:? ?([\d\w]+[ ,-]*[\d?\w?]*?[ ,-]*[\d?\w?]*?[ ,-]*[\d?\w?]*?)',text)[0])
    print(date_list)
    
```
