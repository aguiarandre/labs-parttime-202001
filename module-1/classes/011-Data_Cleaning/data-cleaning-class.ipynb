# Low variance variables
>    - Brings no information
>    - More variance usually means more information 
>    - Percentile

```python
low_variance_cols = []

for col in vehicles._get_numeric_data():
    # if up to the ninety percentile, you are still stuck at the minimum value, remove it.
    minimum = min(data[col])
    ninety_perc = np.percentile(data[col], 90)
    #print(col, minimum, ninety_perc)
    
    if ninety_perc == minimum:
        low_variance.append(col)

print(low_variance_cols)
```

```python
data = data.drop(low_variance, axis=1)
```
