# Another example

<ol>
    <li> request the url </li>
    <li> get its html </li>
    <li> convert to soup </li>
    <li> being a soup, we can look for tags </li>
</ol>
    

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'
```

```python
response = requests.get(url)
```

```python
response
```

```python
html = response.content
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
soup
```

## Find the `table` tag whose `class=sortable wikitable`

```python
table = soup.find_all('table', attrs={'class' : 'sortable wikitable'})[0]
```

## get the table header `<th>`

```python
headers = [header.text.strip() for header in table.find_all('th')]
```

## get the table data `<td>`

```python
data = [data.text.strip() for data in table.find_all('td')]
nrows = int(len(data)/3)
ncols = 3
```

## append results

```python
pd.DataFrame(np.array(data).reshape((nrows, ncols)), columns=headers)
```

# CS GO example

```python
n = 100

url = f'https://www.hltv.org/results?offset={n}'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup
```

```python
websites = ['https://www.hltv.org' + item.find_all('a')[0]['href'] for item in soup.find_all('div', attrs={'class':'result-con'})]
```

```python
websites
```

```python
websites[0]
```

```python
url = websites[0]
soup = BeautifulSoup(requests.get(url).content)
```

```python
from tqdm.auto import tqdm
```

```python
df = pd.DataFrame()

for website in tqdm(websites):
    html = requests.get(website).content
    soup = BeautifulSoup(html)
    
    team1_name = soup.find_all('div', attrs={'class':'team'})[0].find_all('div', attrs={'class':'teamName'})[0].text
    team2_name = soup.find_all('div', attrs={'class':'team'})[1].find_all('div', attrs={'class':'teamName'})[0].text
    team1_score = (soup.find_all('div', attrs={'class':'team'})[0].find_all('div', attrs={'class':'lost'}) + soup.find_all('div', attrs={'class':'team'})[0].find_all('div', attrs={'class':'won'}) + soup.find_all('div', attrs={'class':'team'})[0].find_all('div', attrs={'class':'tie'}))[0].text 
    team2_score = (soup.find_all('div', attrs={'class':'team'})[1].find_all('div', attrs={'class':'lost'}) + soup.find_all('div', attrs={'class':'team'})[1].find_all('div', attrs={'class':'won'}) + soup.find_all('div', attrs={'class':'team'})[1].find_all('div', attrs={'class':'tie'}))[0].text
    
    
    mini_df = pd.DataFrame({'team_a' : [team1_name],
                        'team_b' : [team2_name],
                        'team_a_score': [team1_score],
                        'team_b_score': [team2_score]})
    df = pd.concat([df, mini_df])        
```
