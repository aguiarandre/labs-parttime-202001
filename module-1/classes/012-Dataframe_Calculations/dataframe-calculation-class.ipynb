```python
def normalize_columns(dataframe, is_inplace=False):
    """
    Receive a dataframe and put its columns in lowrercase.
    It also removes spaces for underlines.
    It also removes accents!!!!
    """
    import unidecode
    
    mapper = {column : unidecode.unidecode(column.lower().replace(' ','_')) for column in animals.columns}
    
    return dataframe.rename(columns=mapper, inplace=is_inplace)
```

> Percentiles: 
    - q50 means that 50% of your dataset is below that number. That is the median value.
    - q90 means that 90% of your dataset is below that number.
    - q10 means that 10% of your dataset is below that number.
